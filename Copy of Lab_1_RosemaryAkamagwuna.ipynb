* Write a function to check if a number is prime (use `for` loop and `if` statements)
* Loop through the integers between 2-5000, and store the prime number between 2-5000 into an array, print out the array and size.
* Print out the sum, mean, standard deviation of the array.
* Generate a histogram plot to show the distribution of prime number between 2 and 5000, include axis labels and plot title.


```
import numpy as np
```

```
def is_prime(x):
    if x > 1:
      for i in range(2, x+1):
        if x != i and x % i == 0:
          break
      else:
        return True
    return False
is_prime(3)
```

```
primes_array = np.array([i for i in range (2,5000) if is_prime(i)])
print('Primes Array\n', primes_array)
array_size = primes_array.size
print('Primes Array Size\n', array_size)
array_sum= np.sum(primes_array)
print('Primes Array Sum\n', array_sum)
array_mean= np.mean(primes_array)
print('Primes Array Mean\n', array_mean)
array_std= np.std(primes_array)
print('Primes Array Standard Deviation\n', array_std)
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
mu = array_mean
sigma = array_std
# the histogram of the data
plt.hist(primes_array, 50,facecolor='b', alpha=0.75)

plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Prime Number Distribution Between 2 - 5000')
plt.text(3200, 25, r'$\mu=$'+str(round(array_mean,2)) +'$,\ \sigma=$'+str(round(array_std,2)))
plt.grid(True)
plt.show()
```

* Create a **symmetric** random number array (9 by 9) using Numpy, print out the results
* Calculate the norms of the array (explore different norms using help function), print out the results
* Calculate the determinant, trace, inverse of the matrix, print out the results
* Perform eigen decomposition  for the matrix, print out the eigen vectors and eigen values


```
import numpy as np
nine = np.random.rand(9,9)
nine_symm = (nine + nine.T)/2
print(nine_symm)
```

```
euc_norm = np.linalg.norm(nine_symm) #Euclidean Norm
print('Euclidean Norm of nine_symm\n', euc_norm)
one_norm = np.linalg.norm(nine_symm,1) #1- Norm
print('1-Norm of nine_symm\n', one_norm)
inf_norm = np.linalg.norm(nine_symm,np.inf) #Infinity- Norm
print('Infinity-Norm of nine_symm\n', inf_norm)
```

```
nine_deter = np.linalg.det(nine_symm)
print('Determinant of nine_symm\n', nine_deter)
nine_trac = np.trace(nine_symm)
print('Trace of nine_symm\n', nine_trac)
nine_inv = np.linalg.inv(nine_symm)
print('Inverse of nine_symm\n', nine_inv)
```

```
eigVals,eigVecs=np.linalg.eig(nine_symm) # calculate the eigen value, eigen vectors of the symmetric matrix
print('eigVals\n',eigVals,'\neigVecs\n',eigVecs)

eigVals_diag=np.diag(eigVals)  # create a dignal matrix using eigen values
print('eigVals_diag\n',eigVals_diag)

nine_symm_d=eigVecs.dot(eigVals_diag).dot(eigVecs.T) # calculate the decomposition Q*diag(lamda)*Q_T, A_d should be the same with A
print ('nine_symm_d\n',nine_symm_d)
```

Generate two subplots including following:
* Sigmoid function: $y=1/(1+e^{-x})$, $x$ range is (-5, 5)
* tanh function (numpy.tanh), $x$ range is (-5, 5)


```
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-5 , 5, 10)
y1 = 1/(1 + np.exp(-x1))

x2 = np.linspace(-5 , 5, 10)
y2 = np.tanh(x2)


plt.subplot(2, 1, 1)
plt.plot(x1, y1, '+-')
plt.title('Sigmoid and Tanh Subplots')
plt.ylabel('Sigmoid(x)')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('x')
plt.ylabel('Tanh(x)')

plt.show()

```
